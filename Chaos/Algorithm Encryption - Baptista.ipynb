```python
def chaotic_function(x_n,r=3.9):
    ''' The function is a simulation of chatoic behaviour 
        based on equation  x_n+1 = r * x_n * (1 - x_n)
        where each new_value(x_n+1) is based on old_value(x_n) with a seed vaue r 
    '''
    '''Arguments 
        x_n (previous value and seed value)
        r=custom chatoic constant
    '''
    return r*x_n*(1-x_n)
```

```python
def chaotic_encryption_function(x_n=0.5,r=3.9,x_min=1):
    '''
        this function returns 256 unique values 
            ->which can only be generated for specific  : x_n , r ,x_min 
            ->further used in character mapping
            ->my_mapping_values is most important thing others are used for visulisation purpose only
    '''
    ''''x_n initinal seed value
        r=parameter (constant)
        x_min # used in encyption mapping (logistic mapping)
        chaotic_function is a helper function for this function and is important !
    '''
    my_values=[] ; iterations=[] 
    for i in range(1,x_min+256+500):#X-max tak map poltting
        x_n_1 = chaotic_function(x_n,r)
        my_values.append(x_n_1)
        iterations.append(i)
        x_n=x_n_1
    #at now i have my values and iterations
    if(len(my_values)!=len(iterations)):
        print("This is an error in fn=chaotic_encryption_function , error_code =101")
    my_mapping_values = my_values[x_min:x_min+256]
    if(len(my_mapping_values)!=256):
        print("This is an error in fn=chaotic_encryption_function , error_code = 102")
    return my_mapping_values,my_values,iterations
```

```python
def encrytion_function(text,x_n=0.5,r=3.9,x_min=1):
    my_mapping_values,my_values,iterations=chaotic_encryption_function(x_n,r,x_min)
    #changes to check whether the input parameterse form chaotic behaviour , if allowing the user to set the keys (parameters)
    cipher_text=[]
    for i in range(len(text)):
        char_to_map = text[i]
        for j in range(len(extended_ascii_char)):
            if(char_to_map==extended_ascii_char[j]):
                cipher_text.append(my_mapping_values[j])
                break
    return cipher_text
```
