```python
def chaotic_function(x_n,r=3.9):
    ''' The function is a simulation of chatoic behaviour 
        based on equation  x_n+1 = r * x_n * (1 - x_n)
        where each new_value(x_n+1) is based on old_value(x_n) with a seed vaue r 
    '''
    '''Arguments 
        x_n (previous value and seed value)
        r=custom chatoic constant
    '''
    return r*x_n*(1-x_n)
```

```python
my_values=[] ; iterations=[]
r=3.9
x_n =0.5
for i in range(1,101):#100 iterations
    x_n_1 = chaotic_function(x_n,r)
    my_values.append(x_n_1)
    x_n=x_n_1
    iterations.append(i)
import matplotlib.pyplot as plt
plt.figure(figsize=(14,7))
plt.plot(iterations,my_values,marker="o")
plt.xlabel("Number of iteration (Simulated time series)")
plt.ylabel("Chaotic Value")
plt.title("Chatic Simulation Graph")
plt.grid(True)
plt.show()
```

This graph visualizes the Logistic Map equation:
    x_n+1 = r * x_n * (1 - x_n)
 With:- r = 3.9 (chaotic regime)- x0 = 0.5 (initial condition)- 100 iterations
 Explanation:- The system is deterministic but exhibits unpredictable behavior.- Small changes in x0 lead to completely different sequences.- The output remains bounded (between 0 and 1) but never settles.- At r = 3.9, the system is fully chaotic
