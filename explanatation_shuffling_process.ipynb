```python
# Column-wise sort indices
# np.set_printoptions(threshold=np.inf)
np.set_printoptions(threshold=np.inf, linewidth=200, suppress=True)

import numpy as np
mat=np.random.randint(0,256,(10,10,3))
# print(mat)
for i in range(mat.shape[0]):
    for j in range(mat.shape[1]):
        mat[i][j][1]=i
        mat[i][j][2]=j
print(mat)
```

```python
# Get sorting indices from the first channel for each row
sort_idx = np.argsort(mat[:, :, 0], axis=1)
# Row indices for broadcasting
rows = np.arange(mat.shape[0])[:, None]
# Apply sorting so all 3 channels follow the first channel's order
sorted_mat = mat[rows, sort_idx]
print(sorted_mat)
```
